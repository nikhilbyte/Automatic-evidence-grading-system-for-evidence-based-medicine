```python
fyles = []
abs = []
tit = []
pt = []
for fils in os.listdir("/content/devtestset"):
  with open(os.path.join("/content/devtestset",fils), 'r') as f:
    data = f.read()
  bs_data = BeautifulSoup(data, 'xml')
  fyles.append(fils)
  try:
    abs.append(bs_data.find_all('abstract')[0].get_text())
  except:
    abs.append(bs_data.find_all('abstracttext')[0].get_text())
  tit.append(bs_data.find_all('articletitle')[0].get_text())
  try:
    pt.append(bs_data.find_all('publicationtype')[0].get_text())
  except:
    pt.append("N/A")

  # print(bs_data.find_all('abstract')[0].find_all('abstracttext',attrs={"nlmcategory":"METHODS"}))
  # print(bs_data.find_all('abstract')[0].find_all('abstracttext',attrs={"nlmcategory":"RESULTS"}))
  # print(bs_data.find_all('abstract')[0].find_all('abstracttext',attrs={"nlmcategory":"CONCLUSIONS"}))
```
